```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris

import warnings
warnings.filterwarnings('ignore')
```

```python
df= pd.read_csv('iris.csv')
```

```python
df.head()
```

```python
df=df[["petal length","sepal length"]]
```

```python
df.head(3)
```

```python
model= KMeans(n_clusters=3)
y_model=model.fit_predict(df)
y_model
```

```python
model.labels_
```

```python
plt.scatter(df.iloc[:,0],df.iloc[:,1],c=model.labels_,cmap='viridis'  )
```

```python
pd.value_counts(y_model)
```

```python
#looking at the points which belong to cluster0

df[y_model==0]
```

```python
#looking at the points which belong to cluster1

df[y_model==1]
```

# Selecting no of clusters K

```python
#Th Elbow Method

wcss=[]
K= range(1,15)
for k in K:
    km=KMeans(n_clusters=k)
    km=km.fit(df)
    wcss.append(km.inertia_)
    
plt.plot(K,wcss,marker='x') 
plt.xlabel('k')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
    
```
